```
import io,sqlite3,csv
import pandas as pd

df_Zillow = pd.read_csv('/content/Zip_Zhvi_2bedroom.csv')
df_airbnb = pd.read_csv('/content/listings.csv')

conn = sqlite3.connect("RentalPrice.db")
conn.text_factory = str

df_Zillow.to_sql('Zillow_Details2', conn)
df_airbnb.to_sql('airbnb_details2', conn)
```

```
df_airbnb_notnull = df_airbnb[df_airbnb['zipcode'].notnull()]
```

```
df_airbnb_notnull.shape
```

```
df_airbnb_zip = df_airbnb[df_airbnb['zipcode'].isnull()]

df_airbnb_zip.shape
```

```
def lookup(x):
  try:
    str1 = str(x['latitude']) + ',' + str(x['longitude'])
    print(str1)
    geolocator = Nominatim()
    location = geolocator.reverse(str1)
    return(location.raw['address']['postcode'])
  except:
    pass

df_airbnb_zip['zipcode'] = df_airbnb_zip.apply(lambda row: lookup(row), axis=1)
```

```
df_airbnb_zip = df_airbnb_zip[df_airbnb_zip['zipcode'].notnull()]
```

```
df_airbnb_zip.shape
```

```
df_airbnb_final = pd.concat([
    df_airbnb_zip,df_airbnb_notnull
],ignore_index=True)
```

```
sql_string = '''SELECT a.RegionName as zipcode,a.City,a."2017-06" as latest_sale,b.latitude,b.longitude,b.bedrooms,b."price" as daily_price,
b."cleaning_fee",b."review_scores_rating",b."review_scores_location",b."review_scores_value"
FROM Zillow_Details4 a inner join 
airbnb_details4 b on a.RegionName = b.zipcode
where a.City = 'New York'
'''

df_hdl = pd.read_sql(sql_string, conn)

print("Testing2",df_hdl.shape)
```

```
df_hdl.info()

df_hdl['daily_price'].replace(',','',inplace=True)
```

```
import re

df_hdl['daily_price'] = df_hdl['daily_price'].replace({'\$': '', ',': ''}, regex=True)
```

```
df_hdl['daily_price'] = df_hdl['daily_price'].map(lambda x: x.lstrip(','))
```
